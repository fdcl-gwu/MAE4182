## Bode Plot

Bode plot can be generated by the command [control.bode](https://web.math.princeton.edu/~cwrowley/python-control/generated/control.bode.html#control.bode).

For example, the bode plot of $G(s)=10 s +1$ can be plotted by

```python
import control
import numpy as np
import matplotlib.pyplot as plt

#%matplotlib notebook

sysG = control.tf([10, 1],1)
mag, phase, omega = control.bode(sysG, plot=True)
```

Or the magnitude and the phase at $\omega =0.1$ are computed by

```python
mag, phase, omega = control.bode(sysG, [0.1], plot=False)
print(mag,phase)
```
