# Root Locus

The root locus can be genearted by using the command [control.root_locus](https://python-control.readthedocs.io/en/0.9.4/generated/control.root_locus.html).

First, the characteristic equation for the controlled system should be rearranged into the form of
$$ 1 + K G(s)H(s) = 0.$$

Then, the transfer function corresponding to $G(s)H(s)$ is given to the commnad, `control.root_locus`


## Examples


Consider the following example that we studied in class.

$G(s)H(s) = \dfrac{1}{s\{(s+4)^2+16\}}$



```python
import control
import matplotlib.pyplot as pl
import numpy as np

numGH = 1
denGH = [1, 8, 32, 0]

sysGH = control.tf(1, denGH)

root, K = control.root_locus(sysGH)
```

Next suppose 

$G(s)H(s) = \dfrac{s+1}{s^2}$

```python
sysGH2= control.tf([1, 1], [1, 0, 0])
roots, K = control.root_locus(sysGH2, xlim=(-5, 1), ylim=(-2,2))
```
