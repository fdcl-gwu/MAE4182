## Bode Plot

Bode plot can be generated by the command [control.bode](https://web.math.princeton.edu/~cwrowley/python-control/generated/control.bode.html#control.bode).

For example, the bode plot of $G(s)=\dfrac{s+0.1}{s+10}$ can be plotted by

```python
import control
import numpy as np
import matplotlib.pyplot as plt


sysG = control.tf([1, 0.1],[1, 10])
mag, phase, w = control.bode(sysG, plot=True)
```

Bode plot of $G(s) = \dfrac{2000(s+0.5)}{s(s+10)(s+50)}$

```python
sysG = control.tf( [2000, 1000], [1, 60, 500, 0])
mag, phase, w = control.bode(sysG, plot=True)
```

Bode plot of $G(s) = \dfrac{40s}{s^2 + 0.4s +4}$

```python
sysG = control.tf([40, 0], [1, 0.4, 4])
mag, phase, w = control.bode(sysG, plot=True)
```

Bode plot of $G(s) = \dfrac{10s}{(s+0.1)(s^2+s+100)}$

```python
sysG = control.tf([10, 0], np.polymul([1, 0.1],[1, 1, 100]))

mag, phase, w = control.bode(sysG, plot=True)
```
